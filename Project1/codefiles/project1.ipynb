The Commercial Modular Aero-Propulsion System Simulation (CMAPSS) is a system developed by NASA to study engine degradation, specifically in turbofan jet engines. This system is implemented using MATLAB and Simulink.

A turbofan engine, which is the focus of this study, comprises several key components, including the Fan, Low Pressure Compressor (LPC), High Pressure Compressor (HPC), Low Pressure Turbine (LPT), and High Pressure Turbine (HPT). Each of these components can degrade over time during operation. When the degradation in any component exceeds a certain threshold, the engine is considered to have failed.

To prevent mid-air failures, which are critical given the potential risk to human lives, jet engines are inspected before every take-off. This periodic maintenance is a necessary, though not cost-effective, safety measure. However, there is a need for a more efficient system that can provide early warnings of potential failures.

An effective early warning system can help prepare for or prevent problems before they occur. The goal of a predictive maintenance system is to predict the Remaining Useful Life (RUL) of engine components as accurately as possible. The prediction should not be too early, which could lead to unnecessary maintenance costs, nor too late, which could result in catastrophic failures.

## Classification Problem Formulation

The challenge in this competition is to predict the Remaining Useful Life (RUL) of an engine using data from sensors and operational conditions.

In this project, we will simplify the problem by converting it into a classification task. The classification will categorize the engine's condition into three types: **Good Condition**, **Moderate Condition**, and **Warning Condition**.

### Labels Corresponding to Each Condition

- **Good Condition** - 0
- **Moderate Condition** - 1
- **Warning Condition** - 2

## Defining the Labels

The engine's condition will be defined using the **Life Ratio (LR)**, which is the ratio between the **Current Cycle** and the **End of Life (EOL)** cycle. An **LR = 0** indicates the component has just started its degradation, while **LR = 1** means the component is completely degraded.

- If **LR ≤ 0.6** - **Good Condition**
- If **0.6 < LR ≤ 0.8** - **Moderate Condition**
- If **LR > 0.8** - **Warning Condition**


```python
# Calculate "LR"
raw_data["LR"] = raw_data["Cycle"].div(raw_data["EOL"])
```

```python
raw_data
```

```python
# Drop ID and EOL
raw_data=raw_data.drop(columns=['ID','EOL'])
```

```python
labels=[]
for i in range (0,len(raw_data)):
    if np.array(raw_data["LR"])[i] <= 0.6:
        labels.append(0)
    elif np.array(raw_data["LR"])[i] <= 0.8 :
        labels.append(1)   
    else :
        labels.append(2)  
        
raw_data["labels"]=labels
raw_data  = raw_data.drop(columns=['LR'])   
raw_data
```

<h2>Importing all the Training Data-sets</h2>

```python
df1 = pd.read_csv('Training_1_all_features.csv')
df2 = pd.read_csv('Training_2_all_features.csv')
df3 = pd.read_csv('Training_3_all_features.csv')
df4 = pd.read_csv('Training_4_all_features.csv')
```

**X** is the Feature Matrix and **y** is the label's column

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.10, random_state = 0)
```

<h2>Training a Random Forest model</h2>

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
RF.fit(X_train, y_train)
```

## Randomized Search with Cross-Validation for Hyperparameter Optimization

To address the issue of overfitting observed in the previous models, we need to perform hyperparameter tuning. For this, we will utilize Randomized Search with 3-fold cross-validation to find the optimal set of hyperparameters.

This approach will allow us to efficiently explore a wide range of hyperparameter values and identify the best-performing configuration.


```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators=[200, 400, 300]
# Maximum number of levels in tree
max_depth= [10, 20, 30]
# Max number of samples given to each tree 
max_samples= [0.2,0.3,0.4 ]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               'max_samples': max_samples}
print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, cv = 3,n_jobs=-1)
# Fit the random search model
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
RF_randSer = RandomForestClassifier(n_estimators=300,max_samples=0.4,max_depth= 20)
RF_randSer.fit(X_train, y_train)
y_pred_RF = RF_randSer.predict(X_test)

cm = pd.DataFrame(confusion_matrix(y_test, y_pred_RF, normalize ='true')*100)
colormap = sns.color_palette("Reds")
sns.heatmap(cm, annot=True,cmap=colormap)

accuracy_score(y_test, y_pred_RF)
```
