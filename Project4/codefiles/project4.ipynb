# Prediction on Diabetes Patient's Hospital Readmission

![](https://www.docwirenews.com/wp-content/uploads/2019/05/readmissions.jpg)

Problem Statement and Objective
A hospital readmission is when a patient who is discharged from the hospital, gets re-admitted again within a certain period of time. Hospital readmission rates for certain conditions are now considered an indicator of hospital quality, and also affect the cost of care adversely. For this reason, Centers for Medicare & Medicaid Services established the Hospital Readmissions Reduction Program which aims to improve quality of care for patients and reduce health care spending by applying payment penalties to hospitals that have more than expected readmission rates for certain conditions. Although diabetes is not yet included in the penalty measures, the program is regularly adding new disease conditions to the list, now totaling 6 for FY2018. In 2011, American hospitals spent over $41 billion on diabetic patients who got readmitted within 30 days of discharge. Being able to determine factors that lead to higher readmission in such patients, and correspondingly being able to predict which patients will get readmitted can help hospitals save millions of dollars while improving quality of care. So, with that background in mind, we used a medical claims dataset (description below), to answer these questions:

1. What factors are the strongest predictors of hospital readmission in diabetic patients?
2. How well can we predict hospital readmission in this dataset with limited features?

## Feature Engineering

This is highly subjective and partly depends on knowledge of health care services and understanding the potential relationships between features. There are numerous approaches to explore. I tried some...

* Service utilization: The data includes variables for the number of inpatient admissions, emergency room visits, and outpatient visits for a given patient over the past year. These serve as crude measures of how much hospital/clinic services a person has used in the past year. I created a new variable called "service utilization" by combining these three metrics (see figure below). The objective was to assess which version yields better results. Although I did not apply any special weighting to the components of service utilization, I aimed to try a simple approach at this stage.

```python
df['service_utilization'] = df['number_outpatient'] + df['number_emergency'] + df['number_inpatient']
```

* Number of medication changes: The dataset includes 23 features representing various drugs (or combinations), each indicating whether a change in that medication occurred during the patient's current hospital stay. Previous research has shown that medication changes for diabetics upon admission are associated with lower readmission rates. I decided to count the total number of medication changes for each patient and created a new feature from this count. The rationale was to simplify the model and potentially uncover a relationship with the total number of changes, irrespective of which specific drug was altered.

```python
keys = ['metformin', 'repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride', 'glipizide', 'glyburide', 'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'insulin', 'glyburide-metformin', 'tolazamide', 'metformin-pioglitazone','metformin-rosiglitazone', 'glimepiride-pioglitazone', 'glipizide-metformin', 'troglitazone', 'tolbutamide', 'acetohexamide']
for col in keys:
    colname = str(col) + 'temp'
    df[colname] = df[col].apply(lambda x: 0 if (x == 'No' or x == 'Steady') else 1)
df['numchange'] = 0
for col in keys:
    colname = str(col) + 'temp'
    df['numchange'] = df['numchange'] + df[colname]
    del df[colname]
    
df['numchange'].value_counts()  
```

* Dealing with age: There are various approaches to handling this. The dataset provides age in 10-year categories, so the exact age of each patient is unknown. While previous studies on this dataset used age categories as nominal variables, I aimed to assess the effect of increasing age on readmission, even in a simplified manner. To achieve this, I assumed that the patient’s age is approximately at the midpoint of the age category. For example, if the age category is 20–30 years, I assumed the age to be 25 years. This conversion of age categories to midpoints resulted in a numeric variable.

```python
# code age intervals [0-10) - [90-100) from 1-10
for i in range(0,10):
    df['age'] = df['age'].replace('['+str(10*i)+'-'+str(10*(i+1))+')', i+1)
df['age'].value_counts()
```

Collapsing Multiple Encounters for the Same Patient: Some patients in the dataset had more than one encounter. These could not be counted as independent encounters as this would bias the results towards patients with multiple encounters. Therefore, I employed several techniques to collapse and consolidate multiple encounters for the same patient:

* Considered more than 2 readmissions across multiple encounters as readmission for the collapsed record.
* Calculated the average hospital stay across multiple encounters.
* Considered the percentage of medication changes across multiple encounters.
* Replaced the unique encounter ID with the total number of encounters.
* Combined diagnoses across multiple encounters into a list.

However, for features such as "diagnosis," combining multiple categorical values into an array did not seem meaningful for building the data model. I then examined using the first and last encounters separately as possible representations of multiple encounters. The last encounters resulted in extremely imbalanced data for readmissions (96/4 Readmissions vs. No Readmissions), so I decided to use the first encounters of patients with multiple encounters. This approach reduced the dataset to approximately 70,000 encounters.
